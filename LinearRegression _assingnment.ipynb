```python
import numpy as np
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/drive/MyDrive/Salary_dataset.csv")
df
```

```python
plt.scatter(df.YearsExperience,df.Salary)
plt.xlabel("YearsExperience", color="green")
plt.ylabel("Salary", color="green")
plt.title("Salary Prediction", color="BLUE")
plt.show()
```

```python
lr = linear_model.LinearRegression()
lr.fit(df[['YearsExperience']],df.Salary)
```

```python
lr.predict([[2.5]])
```

```python
lr.score(df[['YearsExperience']],df.Salary)
lr.coef_
```

```python
lr.intercept_
```

```python
predicted = 9449.96232146 *2.5 + 24848.203966523193
print(predicted)
```

```python
plt.scatter(df.YearsExperience,df.Salary)
plt.xlabel("YearsExperience", color="green")
plt.ylabel("Salary", color="green")
plt.plot(df.YearsExperience, lr.predict(df[['YearsExperience']]), color="red")
plt.title("Salary Prediction", color="BLUE")
plt.show()
```
